### Forbes Richest Athletes 1990 - 2020


This Csv data, contains such colums: 


1- "S.NO" - a unique serial number which is assigned to each athlete

2 - "Name" - which represents athletes themselves 

3 - "Nationality" - a nationality of each athlete

4 - "Current Rank" - a rank assigned to ecah athlete in the year specified by the "Year" column 

5 - "Previous Year Rank	" - a rank assigned to ecah athlete in the year previous to the "Year" column 

6 - "Sport" - a type of sport each athlete practies 

7 - "Year" - the deffinition of the current year 

8 - "earnings ($ million)" - the amount of money quantified in millions earned by each athlete 



Each row in this data represnts the column variables named above for each specific athlete name. 

```python
#Import libraries 
import numpy as np
import pandas as pd
import seaborn as sns
import ast, json

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the athlete dataset
athlete_df = pd.read_csv("Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020).csv")
athlete_df.head()
```

### Data Cleaning 

1 - Make sure no infomation in the data is duplicant

2 - Remove any "string" information about the missing data within the integer containing columns (e.g. "?"; "non"; "not stated") and replace it with an emprty sapce. (function - replace())

3 - Use a further function to replace any empty spaces with the data with 0 values. (function - fillna(0))

4 - Check all the integer containing columns for values within them. (function - unique())

5 - Convert integer information into an integer format for furhter visualization use (function - int64)



```python
#Remove any possible duplicate data
athlete_df.drop_duplicates(keep=False)
```

```python
athlete_df["Previous Year Rank"] = athlete_df["Previous Year Rank"].str.replace('>', '')
athlete_df["Previous Year Rank"] = athlete_df["Previous Year Rank"].str.replace('not ranked', '0')
athlete_df["Previous Year Rank"] = athlete_df["Previous Year Rank"].str.replace('none', '0')
athlete_df["Previous Year Rank"] = athlete_df["Previous Year Rank"].str.replace('??', '0')
athlete_df["Previous Year Rank"] = athlete_df["Previous Year Rank"].str.replace('?', '0')
athlete_df
```

```python
#Clean the data of any empty values, such as NaN, and replace them with zeroes 
df = athlete_df[athlete_df['Previous Year Rank']!= ""]
df = athlete_df.fillna(0)
df
```

```python
#Use the astyoe function for the "Previous Year Rank" column as the np.int64 was unable to carry out the conversion 
df["Previous Year Rank"] = df["Previous Year Rank"].astype('int64')
```

### Data Visualization

Represnt data patterns within the data, for easy visual representation and interpretation

1 - Show the top 5 richest athletes within the data (starting form the richest)

2 - Show the top 5 most profitable sport types (starting with the most profitable)

3 - Show the rank difference for 5 athelets (starting with the biggest rank difference)

4 - Create a heatmap to visulay show the correlations between the integer containing data 

5 - Use a pair plot to represent the whole data as a whole (colour coordinate for each athlete within the data)

```python

top_five_richest_athlets = df.sort_values(by = "earnings ($ million)", ascending = False).head()
chart = sns.barplot(
    x = "Name",
    y = "earnings ($ million)", 
    data = top_five_richest_athlets, 
    estimator = sum, 
    errorbar = None,
    color = '#69b3a2')
plt.title('Top Richest Athelets (Forbes)')
```

```python
most_profitable_type_of_sport = df.sort_values(by = "earnings ($ million)", ascending = False).head(5)
chart = sns.barplot(
    x = "Sport",
    y = "earnings ($ million)", 
    data = most_profitable_type_of_sport, 
    estimator = sum, 
    errorbar = None,
    color = '#69b3a2')
plt.title('Most Profitable Sports (Forbes)')
```

```python
difference = df["Current Rank"] - df["Previous Year Rank"]
df["Rank Difference"] = difference
difference_order = df.sort_values(by = "Rank Difference", ascending = False).head()
plt.figure(figsize = (15,5))
chart = sns.barplot(
    x = "Name",
    y = "Rank Difference", 
    data = difference_order, 
    estimator = sum, 
    errorbar = None,
    color = '#69b3a2')
plt.title('Rank Differnce between the "Current Year Rank" and "Previous Year Rank"')
```
